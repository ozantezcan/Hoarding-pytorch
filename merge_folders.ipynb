```python
import os
rootdir='Z:\\Grants\\Image Project\\Data_Sets\\Synthetic\\Rated'
targetdir='\\RuralHome_2017-06-16'
roomdirs=['\\BR','\\Kitchen','\\LR']

fr_count=0
for rm in roomdirs:
    for cir in range(1,10):
        dirname="".join((rootdir,targetdir,rm,'//',str(cir)))
        print(dirname)
        files = os.listdir(dirname)
        for fle in files:
            if fle.endswith('png'):
                fleArr=fle.split('.')
                fleName=fleArr[0]
                targetname="".join((dirname,'//',fle))
                destname="".join((dirname,'//fr',str(fr_count),'.png'))
                os.rename(targetname,destname)
                
                targetname="".join((rootdir,targetdir,'//logs//',fleName,'_log.txt'))
                destname="".join((rootdir,targetdir,'//logs//fr',str(fr_count),'_log.txt'))
                os.rename(targetname,destname)
                fr_count=fr_count+1;
```
